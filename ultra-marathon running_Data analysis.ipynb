###What age groups are the best in the 50mi race (20 + races min)

```python
df3.query('`Event distance/length` == "50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count >= 20')
```

###What age groups are the worst in the 50mi race (20 + races min) (show 20)

```python
df3.query('`Event distance/length` == "50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 19')
```

###Seasons for the data -> Slower in summer than winter

####Spring 3-5

####Summer 6-8

####Fall 9-11

####Split between two decimals


```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season']= df3['race_month'].apply(lambda x: 'Winter' if x >11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```
